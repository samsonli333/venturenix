```python
import requests 
from bs4 import BeautifulSoup

web_site = 'dyson'
def fetch_data(web_site):
    try:
        r = requests.get(f'https://www.price.com.hk/search.php?g=A&q={web_site}&page=1')
        soup = BeautifulSoup(r.text)
        total = soup.find('div',{'class':'pagination-total'})
        total_page = int(total.text.split(' ')[1])
        dict_page = dict()
        for page in range(1,total_page + 1):
            r = requests.get(f'https://www.price.com.hk/search.php?g=A&q={web_site}&page={page}')
            soup = BeautifulSoup(r.text)
            product_list = []
            item_inner_list = soup.findAll('div', {'class': 'item-inner'})
            for item_inner in item_inner_list:
                dict_item_details = dict()
                product_name = item_inner.find('div', {'class': 'line line-01'})
                price = item_inner.findAll('div',{'class':'line line-01 price-range'}) 
                dict_item_details['product_name'] = product_name.text.strip()
                for price_item in price:
                    img = price_item.find('img')
                    if('title' in img.attrs):
                        dict_item_details[img.attrs['title']] = price_item.text.strip()
                    else:
                        dict_item_details['水/行貨'] = price_item.text.strip()
                product_list.append(dict_item_details)
            dict_page[f'Page_{page}'] = product_list
    except:
        print('error')
    finally:
        return dict_page
        

fetch_data(web_site)
```
