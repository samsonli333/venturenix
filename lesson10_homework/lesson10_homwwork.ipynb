## 2. EDA

```python
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].mean())
df_train['age_group'] = df_train['Age'] // 10
```

```python
# option 1
#df_train['Cabin'].fillna('NA').apply(lambda x: x[0])

# option 2
#df_train['Cabin'].str[0]

df_train['cabin_class'] = df_train['Cabin'].str[0].fillna('NA')
```

```python
#Prepare dataset for prediction
# Sex, age, Pclass, sibsp, parch, fare, Cabin[First letter], Survived
# Fill missing age, Cabin
# Use get_dummies on Cabin, Sex
# Optional: bin age

# X -> feature
X = df_train[['Pclass', 'Sex', 'age_group', 'SibSp', 'Parch', 'Fare', 'cabin_class']]
# y -> label
y = df_train['Survived']

X_new = pd.get_dummies(X, columns=['Sex', 'Pclass', 'age_group', 'cabin_class'])


#for i in list(set(list(df_train_s["Cabin_code"]))):
#    df_train_s["Cabin_{}".format(i)] = df_train_s["Cabin_code"].apply(lambda x: 1 if x == i else 0)

#Last step: Put our dataset for prediction


# classifier
clf = LogisticRegression(random_state=0)
clf.fit(X_new, y)



```

```python
# Create x_test
# y_test_predict = clf.predict(X_test)

# x_test -> missing value, mean

df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean())
df_test['age_group'] = df_test['Age'] // 10
df_test['cabin_class'] = df_test['Cabin'].str[0].fillna('NA')

df_test['Fare'] = df_test['Fare'].fillna(df_test['Fare'].mean())
X_test = df_test[['Pclass', 'Sex', 'age_group', 'SibSp', 'Parch', 'Fare', 'cabin_class']]


X_test_new =  pd.get_dummies(X_test,columns=['Sex','Pclass','age_group','cabin_class'])

X_test_new['age_group_8.0'] = False
X_test_new['cabin_class_T'] = False


X_test_new = X_test_new[['SibSp', 'Parch', 'Fare', 'Sex_female', 'Sex_male', 'Pclass_1',
       'Pclass_2', 'Pclass_3', 'age_group_0.0', 'age_group_1.0',
       'age_group_2.0', 'age_group_3.0', 'age_group_4.0', 'age_group_5.0',
       'age_group_6.0', 'age_group_7.0', 'age_group_8.0', 'cabin_class_A',
       'cabin_class_B', 'cabin_class_C', 'cabin_class_D', 'cabin_class_E',
       'cabin_class_F', 'cabin_class_G', 'cabin_class_NA', 'cabin_class_T']]

results = clf.predict(X_test_new)

results
```
